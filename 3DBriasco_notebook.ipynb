```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

# -------------------------UNIVERSIDAD DE 3 DE FEBRERO-------------------------

```python
url_3feb = '/content/drive/MyDrive/Colab_Notebooks/archivos_notebook/3DUN3Feb_transform.csv'
```

# Los datos a analizar son entre las fechas 01/09/2020 al 01/02/2021

```python
df = pd.read_csv(url_3feb)
df
```

```python
plt.style.use('seaborn-whitegrid')
```

# Cantidad de ingresantes por cada carrera


```python
carrera = df['career'].value_counts()
carrera = carrera.to_dict()
fig, ax = plt.subplots(figsize=(12,12))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
parameters = {'axes.labelsize': 25, 'axes.titlesize': 35}
plt.rcParams.update(parameters)
ax.barh(list(carrera.keys()),list(carrera.values()) , color="#199401")
ax.set(title="Ingresantes por carrera", xlabel="Cantidad de ingresantes", ylabel="Carrera")
```

De los cuales

```python
fig = df.groupby(['career','gender']).size().unstack().plot(kind='barh', color=['#9B00B3', '#009B38'], figsize=(15,15))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
parameters = {'axes.labelsize': 25, 'axes.titlesize': 35}
plt.rcParams.update(parameters)
plt.show()
```

# Edad de los ingresantes

```python
edad = df['age'].value_counts()
edad = edad.to_dict()
fig, ax = plt.subplots(figsize=(12,12))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
parameters = {'axes.labelsize': 25, 'axes.titlesize': 35}
plt.rcParams.update(parameters)
ax.bar(list(edad.keys()),list(edad.values()) , color="#1A7A61")
ax.set(title="Ingresantes por carrera", ylabel="Cantidad de ingresantes", xlabel="Edad")
```

```python
fig = df.groupby(['age','gender']).size().unstack().plot(kind='bar', color=['#9B00B3', '#009B38'], figsize=(15,15))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
parameters = {'axes.labelsize': 25, 'axes.titlesize': 35}
plt.rcParams.update(parameters)
plt.show()
```

# Genero de los ingresantes

```python
genero = df['gender'].value_counts()
genero = genero.to_dict()
fig, ax = plt.subplots(figsize=(12,12))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
parameters = {'axes.labelsize': 25, 'axes.titlesize': 35}
plt.rcParams.update(parameters)
ax.bar(list(genero.keys()),list(genero.values()) , color="#7453F0")
ax.set(title="Ingresantes por carrera", ylabel="Cantidad de ingresantes", xlabel="Genero")
```

# --------------------------------COMPARACION ENTRE UNIVERSIDADES-------------------------------------------

```python
df_utn = pd.read_csv(url_utn)
df_3feb = pd.read_csv(url_3feb)

df_utn = df_utn.rename(columns={'career':'utn'})
tabla_utn = df_utn['utn'].value_counts()

df_3feb = df_3feb.rename(columns={'career':'3Feb'})
tabla_3feb = df_3feb['3Feb'].value_counts()

combinacion = pd.concat([tabla_utn, tabla_3feb], axis = 1)
combinacion
```
