```python
import matplotlib.pyplot as plt
import pandas as pd
```

UNIVERSIDAD VILLA MARIA

```python
df = pd.read_csv('datasets/3AUVillaMaria_process.txt', sep='\t')
#print(df['age'].describe())
tamaño = df['age'].size
print(tamaño)
conA = 0
conB = 0
conC = 0
conD = 0
conE = 0
conF = 0
conG = 0
conH = 0
conI = 0
for i in range(tamaño):
    if df['age'].values[i] >= 18 and df['age'].values[i] < 27:
        conA = conA + 1
    if df['age'].values[i] >= 27 and df['age'].values[i] < 36:
        conB = conB + 1
    if df['age'].values[i] >= 36 and df['age'].values[i] < 45:
        conC = conC + 1
    if df['age'].values[i] >= 45 and df['age'].values[i] < 54:
        conD = conD+ 1
    if df['age'].values[i] >= 54 and df['age'].values[i] < 63:
        conE = conE + 1
    if df['age'].values[i] >= 63 and df['age'].values[i] < 72:
        conF = conF + 1
    if df['age'].values[i] >= 72 and df['age'].values[i] < 81:
        conG = conG + 1
    if df['age'].values[i] >= 81 and df['age'].values[i] <= 90:
        conH = conH + 1
```

```python

fig, ax = plt.subplots()

fruits = ['18-27','27-36','36-45','45-54','54-63','63-72','72-81','81-90']
counts = [conA,conB,conC,conD,conE,conF,conG,conH]
bar_labels = ['red', 'blue', '_red', 'orange','red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange','tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('CANTIDAD DE PERSONAS')
ax.set_title('FRANJAS ETARIAS (UNIVERSIDAD VILLA MARIA)')
#ax.legend(title='edades')

plt.show()
```

```python
tamaño = df['gender'].size
print(tamaño)
conA = 0
conB = 0
for i in range(tamaño):
    if df['gender'].values[i] == 'female':
        conA = conA + 1
    if df['gender'].values[i] == 'male':
        conB = conB + 1
```

```python
fig, ax = plt.subplots()

fruits = ['MASCULINO','FEMENINO']
counts = [conA,conB]
bar_labels = ['red', 'blue']
bar_colors = ['tab:red', 'tab:blue']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('CANTIDAD DE PERSONAS')
ax.set_title('GENERO (UNIVERSIDAD VILLA MARIA)')
#ax.legend(title='edades')

plt.show()
```
