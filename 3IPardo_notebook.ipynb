# 1. Universidad Nacional de Jujuy

```python
import pandas as pd
import matplotlib.pyplot as plt

df_jujuy = pd.read_csv("datasets/3IUNJujuy_process.txt", sep='\t')
```

### Cantidad de inscriptos por carrera:

```python
df_career = df_jujuy.pivot_table(index='career', 
                          aggfunc='count', 
                          values='university')
df_career.rename(columns={"university": "count by career"}, inplace=True)
df_career.sort_values(by='count by career', ascending=False, inplace=True)
df_career.reset_index(inplace=True)
df_career
```

```python
career = df_jujuy['career'].value_counts()
career = career.to_dict()
fig, ax = plt.subplots(figsize=(12,12))
ax.barh(list(career.keys()),list(career.values()) , color="#199401")
ax.set(title="INGRESANTES POR CARRERA (UNIVERSIDAD DE JUJUY)", xlabel="CANTIDAD DE INGRESANTES", ylabel="CARRERA")
plt.show()
```

### Edad promedio por carrera:

```python
df_age = df_jujuy.pivot_table(index='career', 
                          aggfunc='mean', 
                          values='age')
df_age.rename(columns={"age": "mean age"}, inplace=True)
df_age.sort_values(by='mean age', ascending=False, inplace=False)
df_age.reset_index(inplace=True)
df_age.style.format(precision=1)
```

### Cantidad de hombres y mujeres por carrera:

```python
df_gen = df_jujuy.groupby(["gender", 'career'])['university'].count()
df_gen = df_gen.to_frame()
df_gen.reset_index(inplace=True)
df_gen.rename(columns={"university": "count_by_career"}, inplace=True)
df_gen.sort_values(by=["gender", "count_by_career"], ascending=False, inplace=False)
df_gen
```

### Cantidad de personas inscriptas por localidad:

```python
df_loc = df_jujuy.pivot_table(index='location', 
                          aggfunc='count', 
                          values='university')
df_loc.rename(columns={"university": "count by location"}, inplace=True)
df_loc.sort_values(by='count by location', ascending=False, inplace=True)
df_loc.reset_index(inplace=True)
df_loc
```
